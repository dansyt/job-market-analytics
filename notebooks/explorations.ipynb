```python
from playwright.sync_api import sync_playwright

url = "https://jobseeker.kalibrr.com/c/pt-akhdani-reka-solusi/jobs/268536/it-support-yogyakarta-shifting"

with sync_playwright() as p:
    browser = p.chromium.launch(headless=False)  # tampilkan browser
    page = browser.new_page()
    page.goto(url)

    page.wait_for_timeout(5000)  # tunggu 5 detik

    html = page.content()

    print("IT Support" in html)
    print(html[:1000])

    browser.close()
```

```python
from bs4 import BeautifulSoup
import json

soup = BeautifulSoup(html, "html.parser")

scripts = soup.find_all("script", type="application/ld+json")

print(len(scripts))
```

```python
data = json.loads(scripts[0].get_text())
print(data)
```
